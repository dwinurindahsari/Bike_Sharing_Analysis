```python
import pandas as pd
```

```python
df = pd.read_csv("day.csv")
df.sample(200)
```

```python
df.info()
```

```python
df.describe()
```

#2. Cleaning


```python

df.isnull().sum()
df.duplicated().sum()
df['dteday'] = pd.to_datetime(df['dteday'])
df.info()
```

#3.Analisis Sederhana

### Rata-rata penggunaan per hari

```python
df.groupby('weekday') ['cnt'].mean()
```

###Rata-rata penggunaan per season

```python
df.groupby('season') ['cnt'].mean()
```

### weather impact


```python
df.groupby('weathersit') ['cnt'].mean()
```

###Rata-rata penggunaan per Bulan

```python

df.groupby('mnth') ['cnt'].mean()
```

###Rata-rata penggunaan per Tahun

```python
df.groupby('yr') ['cnt'].mean()
```

#4. Visualisasi

```python
import matplotlib.pyplot as plt

df.groupby('weekday')['cnt'].mean().plot(kind='bar')
plt.title('Rata-rata penggunaan sepeda per Hari')
plt.show()
```

```python
df.groupby('weathersit')['cnt'].mean().plot(kind='bar')
plt.title('Rata-rata penggunaan sepeda per Cuaca')
plt.show()
```

### Weather
- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds



```python
df.groupby('season')['cnt'].mean().plot(kind='bar')
plt.title('Rata-rata penggunaan sepeda per Season')
plt.show()
```

### season
1 = Spinger
2 = Summer
3 = Fall
4 = Winter


```python
df.groupby('mnth')['cnt'].mean().plot(kind='bar')
plt.title('Rata-rata penggunaan sepeda per Bulan')
plt.show()
```

```python
df.groupby('yr')['cnt'].mean().plot(kind='pie')
plt.title('Rata-rata penggunaan sepeda per Tahun')
plt.show()
```
